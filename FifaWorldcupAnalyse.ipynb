# **PREPROCESSING THE DATA & CREATING FUNCTIONS FOR PLOT**

```python
fix_name = {"possession team1":"possession1",
            "possession team2":"possession2",
            "number of goals team1":"goals1",
            "number of goals team2":"goals2",
            "total attempts team1":"attempts1",
            "total attempts team2":"attempts2",
            "conceded team1":"conceded1",
            "conceded team2":"conceded2",
            "on target attempts team1":"target1",
            "on target attempts team2":"target2",
            "yellow cards team1":"yellowcard1",
            "yellow cards team2":"yellowcard2",
            "penalties scored team1":"penalties1",
            "penalties scored team2":"penalties2",
            "goal preventions team1":"goal_prevention1",
            "goal preventions team2":"goal_prevention2",
            "defensive pressures applied team1":"defensive_pressure1",
            "defensive pressures applied team2":"defensive_pressure2",
            "own goals team1":"own_goals1",
            "own goals team2":"own_goals2",
            "passes team1":"passes1",
            "passes team2":"passes2",
            "passes completed team1":"passes_completed1",
            "passes completed team2":"passes_completed2",
            "red cards team1":"redcards1",
            "red cards team2":"redcards2",
            "offsides team1":"offside1",
            "offsides team2":"offside2",
            "free kicks team1":"freekicks1",
            "free kicks team2":"freekicks2"

            }
df = df.rename(fix_name, axis = 1)
df['possession1'] = df['possession1'].str[:-1].astype('float64')
df['possession2'] = df['possession2'].str[:-1].astype('float64')
df['possession in contest'] = df['possession in contest'].str[:-1].astype('float64')

class BarPlott:
    cntry = set()
    ls = df['team1'].unique().tolist() + df['team2'].unique().tolist()
    for c in ls:
      cntry.add(c)
      countries = list(cntry)
    TotalFactListt = {}
    TotalFactList = []
    TotalFactListt1 = {}
    TotalFactList1 = []
    def Team(self,tname):
      t1g = df[df["team1"] == tname]
      t2g = df[df["team2"] == tname]
      frames = [t1g,t2g]
      self.t12g = pd.concat(frames)
      return self.t12g

    def Total(self,tname):
        TotalFact = []
        x=self.fact+"1"
        y=self.fact+"2"
        for index, row in self.Team(tname).iterrows():
            if row["team1"] == tname:
                TotalFact.append(row[x])
            if row["team2"] == tname:
                TotalFact.append(row[y])
        return TotalFact

    def TotalAvg(self,tname):
        TotalFact = []
        x=self.fact1+"1"
        y=self.fact1+"2"
        for index, row in self.Team(tname).iterrows():
            if row["team1"] == tname:
                TotalFact.append(row[x])
            if row["team2"] == tname:
                TotalFact.append(row[y])
        return TotalFact

    def TotalList(self,fact):
        self.fact = fact
        for i in self.countries:
            self.TotalFactListt[i] = self.Total(i)
        self.TotalFactList = [[c, sum(g)] for c, g in self.TotalFactListt.items()]
        return self.TotalFactList

    def Avg(self,tname):
      return round(sum(self.TotalAvg(tname)) / len(self.TotalAvg(tname)),1)

    def AvgList(self,fact1):
        self.fact1 = fact1
        for i in self.countries:
            self.TotalFactListt1[i] = self.Avg(i)
        self.TotalFactList1 = [[c,g] for c, g in self.TotalFactListt1.items()]
        return self.TotalFactList1

    def BarPlot(self, lst, titl, widthh=1.0):
        import matplotlib.pyplot as plt
        import pandas as pd

        y = pd.DataFrame(lst)
        plt.style.use("dark_background")
        plt.barh(y[0], y[1], color='skyblue', edgecolor='g')
        plt.title(titl)
        plt.gca().invert_yaxis()
        plt.xlim(0, max(y[1]) + widthh)
        for i, value in enumerate(y[1]):
            plt.text(value + 0.1, i, str(value), va='center', ha='left')
        plt.show()

    def TeamNames(self,gpname):
        gt = df[df['category'] == gpname]
        tnl = pd.concat([gt['team1'], gt['team2']]).unique().tolist()
        self.dft = pd.DataFrame(tnl,columns=[gpname])
        return self.dft


    def GroupwiseTeamsPlot(self):
      import pandas as pd
      import matplotlib.pyplot as plt
      # Create DataFrames
      df1 = self.TeamNames("Group A")
      df2 = self.TeamNames("Group B")
      df3 = self.TeamNames("Group C")
      df4 = self.TeamNames("Group D")
      df5 = self.TeamNames("Group E")
      df6 = self.TeamNames("Group F")
      df7 = self.TeamNames("Group G")
      df8 = self.TeamNames("Group H")
      df9 = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8],axis=1)

      fig, ax = plt.subplots(figsize=(15, 3))
      plt.style.use('default')

      table = ax.table(cellText=df9.values, colLabels=df9.columns, cellLoc='center', loc='center')
      table.auto_set_font_size(False)
      table.set_fontsize(12)
      table.scale(1.2, 1.2)
      ax.set_title("GROUPWISE TEAMS", fontsize=16, weight='bold')

      for key, cell in table.get_celld().items():
        cell.set_height(0.1)

      ax.axis('off')


      plt.tight_layout()
      return plt.show()

    def TopItem(self, level, dataa):
        return sorted(dataa, key=lambda item: item[1], reverse=True)[:level]

    def LowestItem(self, level, dataa):
        return sorted(dataa, key=lambda item: item[1])[:level]

def final():
      import pandas as pd
      import matplotlib.pyplot as plt

      fd = df[df['category'] == "Final"]

      teamm1 = {"ARGENTINA":[fd["goals1"].item(),fd["target1"].item(),fd["attempts1"].item(),fd["passes_completed1"].item(),fd["fouls against team2"].item(),fd["yellowcard1"].item(),fd["redcards1"].item()]}
      cate = {"STATS":["Goals","Shot On Target","Attempts","Passes Completed","Fouls Committed","Yellow Cards","Red Cards"]}
      teamm2 = {"FRANCE":[fd["goals2"].item(),fd["target2"].item(),fd["attempts2"].item(),fd["passes_completed2"].item(),fd["fouls against team1"].item(),fd["yellowcard2"].item(),fd["redcards2"].item()]}


      df1 = pd.DataFrame(teamm1)
      df2 = pd.DataFrame(cate)
      df3 = pd.DataFrame(teamm2)


      df4 = pd.concat([df1,df2,df3],axis=1)
      fig, ax = plt.subplots(figsize=(6, 4))
      table = ax.table(cellText=df4.values, colLabels=df4.columns, cellLoc='center', loc='center',)
      table.auto_set_font_size(False)
      table.set_fontsize(12)
      table.scale(1.2, 1.2)
      ax.axis('off')
      for key, cell in table.get_celld().items():
        cell.set_linewidth(0)
      plt.title("FINAL BATTLE", fontsize=16, weight='bold')
      plt.text(0.5, 0.1, 'ARGENTINA (4) 3 - 3 (3) FRANCE', horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, fontsize=14, color='black', fontweight='bold')
      plt.style.use('default')
      return plt.show()
```
